## NLP Data Cleaning

```python
#remove punctuation

import string
string.punctuation
```

```python
def no_punctuation(text):
    nopunct=[words for words in text if words not in string.punctuation]
    words_without_punct=''.join(nopunct)
    return words_without_punct
```

```python
#remove punctuation and lowercase words in 'category,' 'answer,' and 'question'

df['category'] = df['category'].apply(lambda x: no_punctuation(x).lower())
df['answer'] = df['answer'].apply(lambda x: no_punctuation(x).lower())
df['question'] = df['question'].apply(lambda x: no_punctuation(x).lower())
```

```python
df
```

```python
#remove numerals from 'category,' 'answer,' and 'question'

df['category'] = df['category'].str.replace('\d+', '')
df['answer'] = df['answer'].str.replace('\d+', '')
df['question'] = df['question'].str.replace('\d+', '')
```

```python
df
```

```python
#tokenize words in 'category,' 'answer,' and 'question'

df['tokenized_category'] = df['category'].apply(word_tokenize)
df['tokenized_answer'] = df['answer'].apply(word_tokenize) 
df['tokenized_question'] = df['question'].apply(word_tokenize)
```

```python
df
```

```python
#remove stopwords in 'tokenized_category,' 'tokenized_answer,' and 'tokenized_question'

stop = stopwords.words('english')
df['tokenized_category'] = df['tokenized_category'].apply(lambda x: [item for item in x if item not in stop])
df['tokenized_answer'] = df['tokenized_answer'].apply(lambda x: [item for item in x if item not in stop])
df['tokenized_question'] = df['tokenized_question'].apply(lambda x: [item for item in x if item not in stop])
```

```python
df
```

```python
#lemmatize tokenized categories

lemmatizer = WordNetLemmatizer()
df['tokenized_category'] = df['tokenized_category'].apply(lambda row:[lemmatizer.lemmatize(word) for word in row])
df['tokenized_answer'] = df['tokenized_answer'].apply(lambda row:[lemmatizer.lemmatize(word) for word in row])
df['tokenized_question'] = df['tokenized_question'].apply(lambda row:[lemmatizer.lemmatize(word) for word in row])
```

```python
df
```

## Modeling

```python
df
```

```python
#Use 'answer' (Jeopardy questions) as the feature and 'value_adjusted' as the target
X = df['answer']
y = df['value_adjusted']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state=54)
```

```python
# Instantiate the vectorizer

tfidf = TfidfVectorizer()

# Fit the vectorizer on X_train and transform it
X_train_vectorized = tfidf.fit_transform(X_train)
```

```python
# Evaluate the classifier on X_train_vectorized and y_train
baseline_model = MultinomialNB()
baseline_cv = cross_val_score(baseline_model, X_train_vectorized, y_train)
baseline_cv
```

```python
print("Baseline:         ", baseline_cv.mean())
```

```python
y_train.value_counts(normalize=True)
```

This is slightly better than simply guessing the category with the most values (1) but not by much.

```python
baseline_model.fit(X_train_vectorized, y_train)
sns.set_style("white")
plot_confusion_matrix(baseline_model, X_train_vectorized, y_train);
```

```python
#Run model again with stopwords removed

tfidf_stopwords = TfidfVectorizer(stop_words='english')

# Fit the vectorizer on X_train and transform it
X_train_vectorized_stopwords = tfidf_stopwords.fit_transform(X_train)
```

```python
# Evaluate the classifier on X_train_vectorized and y_train
stopwords_removed_model = MultinomialNB()
stopwords_removed_cv = cross_val_score(stopwords_removed_model, X_train_vectorized_stopwords, y_train)
stopwords_removed_cv
```

```python
print("Baseline:         ", baseline_cv.mean())
print("Stopwords Removed:", stopwords_removed_cv.mean())
```

```python
stopwords_removed_model.fit(X_train_vectorized_stopwords, y_train)
plot_confusion_matrix(stopwords_removed_model, X_train_vectorized_stopwords, y_train);
```

That actually made our score lower!

```python
#Try out a regression model

baseline_regression_model = LogisticRegression(random_state=54, solver='saga', max_iter=400)
baseline_regression_cv = cross_val_score(baseline_regression_model, X_train_vectorized, y_train)
baseline_regression_cv
```

```python
print("Baseline:         ", baseline_cv.mean())
print("Stopwords Removed:", stopwords_removed_cv.mean())
print("Regression Baseline:", baseline_regression_cv.mean())
```

```python
baseline_regression_model.fit(X_train_vectorized, y_train)
plot_confusion_matrix(baseline_regression_model, X_train_vectorized, y_train);
```
